```python
import keras
from keras.applications.xception import Xception
# from keras.preprocessing import image
from keras.models import Model
from keras.layers import Dropout, Dense, Activation
```

```python
num_class = 256 # caltech 256
# imagenet input sizse (299,299,3)
weight_decay=1e-4
```

```python
base_model = Xception(weights='imagenet')
base_model.summary()
```

```python
base_model.layers.pop()
base_model.summary()
print(base_model.output) # not changed after layers.pop() is called
```

```python
print(base_model.layers[-1].output, base_model.layers[-1].name)
print(base_model.layers[0].input_shape)
print(base_model.name)
```

```python
x = base_model.layers[-1].output
x = Dropout(0.5)(x)

logits = Dense(num_class, kernel_regularizer= keras.regularizers.l2(weight_decay), name='logits')(x)
probabilities = Activation('softmax')(logits)

model = Model(base_model.input, probabilities, name='xception')
```

```python
model.summary()
print (model.name)
```
